# Data preprocessing

```python
import missingno as mo
```

```python
mo.bar(A) #finding out missing vaalue in dataset in graphical form using missingno library
```

```python
A.isna().sum().plot(kind='barh')
```

```python
#separating the categorical and continuous column from the dataset
cat = []
con = []
for i in A.columns:
    if(A[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
cat #categorical columns
```

```python
con #continuous columns
```

```python
#treating missing values
def replacer(df):
    for i in con:
        x = df[i].mean()
        df[i] = df[i].fillna(x)
        
    for i in cat:
        x = df[i].mode()[0]
        df[i] = df[i].fillna(x)
```

```python
replacer(A)
```

```python
A.isna().sum() #All null values treated
```

```python
for i in A.columns:
    if(A[i].dtypes == "object"):
        sb.boxplot(A[i],A.Width)#Boxplot between categorical and continuous columns
        plt.xlabel(i)
        plt.ylabel("Width")
        plt.show()
        
    else:
        plt.scatter(A[i],A.Width)#Scatter plot between continuous and continuous columns
        plt.xlabel(i)
        plt.ylabel("Width")
        plt.show()
```
